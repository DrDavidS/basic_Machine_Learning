## 参数更新

神经网络的学习目的是找到使损失函数的值尽可能小的参数。这是一个寻找最优参数的问题，解决这个问题的过程就叫**最优化**（optimization）。但是神经网络的优化问题非常困难，因为参数空间庞大且复杂，难以找到最优解。

在之前，我们为了找到最优参数，将参数的梯度（导数）作为线索。使用参数的梯度，沿着梯度方向更新参数，然后重复这个步骤多次，从而逐渐靠近最优参数，这个过程就是随机梯度下降法（Stochastic Gradient Descent），简称 SGD。

> 在Pytorch中，SGD 方法为 `torch.optim.SGD()`，具体文档参见[torch.optim.SGD](https://pytorch.org/docs/stable/optim.html?highlight=sgd#torch.optim.SGD)。

SGD 方法简单，比起胡乱搜索参数空间，总归是个聪明方法。但是随着问题的不同，我们也有比 SGD 更聪明的方法。

我们会简单讲讲 SGD 的缺点，然后介绍其他的最优化方法。

### SGD

首先我们复习一下 SGD。用数学式子表示 SGD 就是：

$$\large \textbf{W}\leftarrow\textbf{W}-\eta\frac{\partial L}{\partial\textbf{W}} \tag{6.1}$$

其中，需要更新的权重参数记为 $\textbf{W}$，把损失函数关于 $\textbf{W}$ 的梯度记为 $\cfrac{\partial L}{\partial\textbf{W}}$。

$\eta$ 表示学习率，是一个**超参数**，实际上会取 $0.01$ 或者 $0.001$ 等实现决定好的值。

式子中的 $\leftarrow$ 表示用右边的值更新左边的值。

如式 6.1 所示，SGD 是朝着梯度方向只前进一定距离（学习率）的简单方法。

为了后续演示方便。首先我们定义神经网络如下，由三个线性层组成：

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    """定义神经网络Net"""
    def __init__(self):
        """初始化各个层"""
        super(Net, self).__init__()
        self.fc1 = nn.Linear(in_features=28*28, out_features=100)
        self.fc2 = nn.Linear(in_features=100, out_features=50)
        self.fc3 = nn.Linear(in_features=50, out_features=10)

    def forward(self, x):
        """定义前向传播"""
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.fc3(x)
        return F.log_softmax(x, dim=1)
```

```python
model = Net()
```

在 PyTorch 中，SGD 被封装在 `optim` 类中，调用方法如下：

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
print(optimizer)
```

通常编程中，变量名 **optimizer** 表示“优化器” 的意思，这里由 SGD 来承担这个角色。参数的更新由 optimizer 负责完成。我们在这里需要做的只是将参数和梯度信息传给 optimizer。

使用类封装，功能的模块化变得更简单。以后我们如果需要使用其他优化器，就把这里的 SGD 换一换就行了。

### SGD 的缺点

虽然 SGD 简单，而且容易实现，但是在解决某些问题的时候可能没什么效率。

这里，我们首先来思考一下，求这个函数的最小值的问题。

$$\large f(x,y)=\frac{1}{20}x^2+y^2 \tag{6.2}$$

```python
def func_old(x, y):
    """式6.2"""
    return x**2 / 20 + y**2
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
import numpy as np

fig = plt.figure()
ax = Axes3D(fig)

x = np.arange(-10, 10, 0.5)  # x 
y = np.arange(-10, 10, 0.5)  # y
X, Y = np.meshgrid(x, y)  # 二维数组生成
Z = func_old(X, Y)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f(x,y)$')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')


# 中文标题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('函数 $f(x,y)$ 的图像\n', size=15)

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.contour(X, Y, Z, levels=50, linewidths=0.5, linestyles='dashdot')  # 绘制等高线

# 中文标题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('函数 $f(x,y)$ 的等高线\n', size=15)

plt.show()
```

如上方绘图所示，式 6.2 中的函数是沿 $x$ 轴方向延伸的山谷状的函数。其等高线呈现向 $x$ 轴方向延伸的椭圆状。

接下来我们分析一下式 6.2 所表示的函数的梯度。如果用图表示梯度，那么如下图所示。

```python
def func(x):
    """式6.2，改为接受一个数组"""
    return ((x[0]**2) / 20) + (x[1]**2)

def _numerical_gradient_no_batch(f, x):
    """
    计算梯度
    
    输入：
    f：函数
    x：数组，多元变量。
    """
    h = 1e-4  # 0.0001
    grad = np.zeros_like(x)  # 生成一个和x形状一样的全为0的数组
    
    for idx in range(x.size):
        tmp_val = x[idx]
        x[idx] = float(tmp_val) + h
        fxh1 = f(x)  # f(x+h)
        
        x[idx] = tmp_val - h 
        fxh2 = f(x)  # f(x-h)
        grad[idx] = (fxh1 - fxh2) / (2*h)
        
        x[idx] = tmp_val  # 还原值
        
    return grad

def numerical_gradient(f, X):
    """计算梯度矢量"""
    if X.ndim == 1:
        return _numerical_gradient_no_batch(f, X)
    else:
        grad = np.zeros_like(X)
        
        for idx, x in enumerate(X):
            grad[idx] = _numerical_gradient_no_batch(f, x)
        
        return grad

x = np.arange(-10, 10, 0.5)  # x 
y = np.arange(-10, 10, 0.5)  # y
X, Y = np.meshgrid(x, y)  # 二维数组生成

X = X.flatten()
Y = Y.flatten()

grad = numerical_gradient(func, np.array([X, Y]).T).T
plt.figure(figsize=(16, 10))
plt.quiver(X, Y, -grad[0], -grad[1],  angles="xy",color="#666666")
plt.xlim([-10, 10])
plt.ylim([-5, 5])
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()

# 中文标题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('函数 $f(x,y)$ 的梯度\n', size=15)

plt.draw()
plt.show()
```

>上面的代码复用了我们在 [3.02 神经网络的训练](https://github.com/DrDavidS/basic_Machine_Learning/blob/master/%E6%B7%B1%E5%BA%A6%E5%AD%A6%E4%B9%A0%E5%9F%BA%E7%A1%80/3.02%20%E7%A5%9E%E7%BB%8F%E7%BD%91%E7%BB%9C%E7%9A%84%E8%AE%AD%E7%BB%83.ipynb) 中编写的绘图代码，看起来比较复杂。不过没关系，我们只看图就好。

如上图所示，这个梯度的特征是， $y$ 轴方向上偏大，$x$ 轴方向上偏小。参照之前的山谷状图片说就是， $y$ 轴方向上坡度大，而 $x$ 轴方向上坡度小。

需要注意的是，虽然我们知道式 6.2 的最小值在 $(x, y) = (0, 0)$ 处，但是在上图中，很多梯度方向并没有指向 $(0, 0)$ 这个点，而是整体地指向了 $y=0$ 这条线。

现在，如果我们对这种形状的函数应用 SGD，从 $(x, y)=(-7.0, 2.0)$ 开始搜索，结果如下：

```python
def gradient_descent(f, init_x, lr=0.5, step_num=100):
    """
    梯度下降法
    
    f：要进行最优化的参数
    init_x：初始值
    lr：学习率，默认为0.5
    step_sum：梯度下降法重复的次数
    """
    x = init_x
    x_history = []  # 保存每一步的信息

    for i in range(step_num):
        x_history.append( x.copy() )

        grad = numerical_gradient(f, x)  # 计算梯度矢量
        x -= lr * grad

    return x, np.array(x_history)

init_x = np.array([-7.0, 2.0])  # 初始位置
lr = 0.9                        # 学习率
step_num = 50                   # 步数

x, x_history = gradient_descent(func, init_x, lr=lr, step_num=step_num)

step = 0.1
x = np.arange(-10, 10, step)
y = np.arange(-10, 10, step)
X, Y = np.meshgrid(x, y)  # 建立网格
Z = func_old(X, Y)

plt.figure(figsize=(10,6))
plt.contour(X, Y, Z, levels=50, linewidths=0.5, linestyles='dashdot')  # 绘制等高线

plt.plot(x_history[:,0], x_history[:,1], '-o')  # 绘制梯度下降过程

plt.xlim(-10, 10)
plt.ylim(-4, 4)
plt.xlabel("$x$")
plt.ylabel("$y$")

# 中文标题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('基于 SGD 的最优化更新路径，呈现之字形朝最小值 $(0,0)$ 移动，效率低\n', size=15)

plt.show()
```

在上图中，SGD 呈现“之”字形移动。这个路径相对是很低效的。

> 通过设置较小的学习率可能在一定程度上避免这个问题。但是学习率较小，则收敛会变慢。

也就是说，SGD 的缺点是，如果函数的形状**非匀向**（anisotropic），比如呈现延伸状，搜索的路径就会变得效率低下。

SGD 低效的**根本原因**是：梯度的方向没有指向最小值的方向。

为了改正 SGD 的缺点，我们接下来会介绍一些替代 SGD 的最优化方法。

### Momentum

Momentum 是“动量”的意思，在初中我们应该就在物理课上学过这个概念了。

用数学式表示 Momentum 方法如下：

$$\large \textbf{v} \leftarrow \alpha\textbf{v} - \eta\frac{\partial L}{\partial \textbf{W}} \tag{6.3}$$

$$\large \textbf{W} \leftarrow \textbf{W} + \textbf{v} \tag{6.4}$$

和之前的 SGD 一样，需要更新的权重参数记为 $\textbf{W}$，把损失函数关于 $\textbf{W}$ 的梯度记为 $\cfrac{\partial L}{\partial\textbf{W}}$。

$\eta$ 表示学习率，式子中的 $\leftarrow$ 表示用右边的值更新左边的值。

这里有一个新变量 $\textbf{v}$，对应物理学上的速度。

式 6.3 表示物体在梯度方向上的受力。在这个力的作用下，物体的速度增加**动量**这一物理量。

式 6.3 中有 $\alpha \textbf{v}$ 这一项。在物体不受任何力时，该项承担物体逐渐减速的任务（比如把 $\alpha$ 设为0.4），类比物体在地面的摩擦力或者空气阻力之类。

```python
def Momentum(f, init_x, lr=0.5, step_num=100, momentum=0.9):
    """
    Momentum 方法
    
    f：要进行最优化的参数
    init_x：初始值
    lr：学习率，默认为0.5
    step_sum：梯度下降法重复的次数
    momentum：模拟摩擦力
    """
    x = init_x
    x_history = []  # 保存每一步的信息
    v = 0 # 一开始v为0
    
    for i in range(step_num):
        x_history.append( x.copy() )

        grad = numerical_gradient(f, x)  # 计算梯度矢量
        v = momentum * v - lr * grad
        x = x + v

    return x, np.array(x_history)

init_x = np.array([-7.0, 2.0])  # 初始位置
lr = 0.9                        # 学习率
step_num = 50                   # 步数
momentum = 0.3                  # Momentum

x, x_history = Momentum(func, init_x, lr=lr, step_num=step_num, momentum=momentum)

step = 0.1
x = np.arange(-10, 10, step)
y = np.arange(-10, 10, step)
X, Y = np.meshgrid(x, y)  # 建立网格
Z = func_old(X, Y)

plt.figure(figsize=(10,6))
plt.contour(X, Y, Z, levels=50, linewidths=0.5, linestyles='dashdot')  # 绘制等高线

plt.plot(x_history[:,0], x_history[:,1], '-o')  # 绘制梯度下降过程

plt.xlim(-10, 10)
plt.ylim(-4, 4)
plt.xlabel("$x$")
plt.ylabel("$y$")

# 中文标题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('基于 Momentum 的最优化更新路径\n', size=15)

plt.show()
```

在上图中，更新的路径就想小球在碗里面滚动一样。

和 SGD 相比，很明显发现这个“之”字形的“程度”减轻了。

这是因为在 $x$ 轴方向上收到的力非常小，但是一直在同一个方向受力，所以朝同一个方向有加速度。

反过来，虽然在 $y$ 轴方向上受到的力很大，但是因为交互地受到正方向和反方向的力，它们会互相抵消，所以 $y$ 轴方向上的速度不稳定，衰减较快。、

因此，和 SGD 的情况相比，可以更快朝 x 轴方向靠近，减弱“之”字形的变动程度。

> 在 PyTorch 中，Momentum 的实现包含在了 `torch.optim.SGD` 中。默认参数 `momentum=0`，即普通的 SGD 方法，可以调整动量参数，让它变成带动量的 SGD 方法。参考文档：[SGD](https://pytorch.org/docs/stable/optim.html?highlight=sgd#torch.optim.SGD)。

### AdaGrad

在神经网络的学习中，学习率（这里记为 $\eta$）的值很重要。学习率过小，会导致花费过多时间；反过来，学习率过大，会导致学习发散而不能正确进行。

在关于学习率的有效技巧中，有一种称为学习率衰减（learning rate decay）的方法，即随着学习的进行，使学习率逐渐减小。实际上，一开始“多”学，然后逐渐“少”学的方法，在神经网络的学习中经常被使用。

逐渐减小学习率，相当于将“全体”参数的学习率一起降低。而 AdaGrad 方法进一步发展了这个想法，针对“一个一个”的参数，赋予“定制”的值。

AdaGrad 会为参数的每个元素适当地调整学习率，与此同时进行学习。用数学式表示就是：

$$\large \textbf{h} \leftarrow \textbf{h} + \frac{\partial L}{\partial \textbf{W}} * \frac{\partial L}{\partial \textbf{W}} \tag{6.5}$$

$$\large \textbf{W} \leftarrow \textbf{W} - \eta\frac{1}{\sqrt{h}} \frac{\partial L}{\partial \textbf{W}} \tag{6.6}$$

和 SGD 一样，需要更新的权重参数记为 $\textbf{W}$，把损失函数关于 $\textbf{W}$ 的梯度记为 $\cfrac{\partial L}{\partial\textbf{W}}$，$\eta$ 表示学习率。

这里有一个新变量 $\textbf{h}$，如式 6.5 所示，它保存了以前所有梯度值的平方和。符号 $*$ 表示对于矩阵元素的乘法，也就是**哈达玛积**(Hadamard product)。

在更新参数时，通过乘以 $\cfrac{1}{\sqrt{h}}$，就可以调整学习的尺度。这意味着，参数的元素中变动较大（被大幅更新）的元素学习率会变小。

也就是说，可以按参数的元素进行学习率衰减，使变动大的参数的学习率逐渐减小。

具体代码如下：

```python
def AdaGrad(f, init_x, lr=0.5, step_num=100):
    """
    AdaGrad 方法
    
    f：要进行最优化的参数
    init_x：初始值
    lr：学习率，默认为0.5
    step_sum：梯度下降法重复的次数
    """
    x = init_x
    x_history = []  # 保存每一步的信息
    h = 0 # 一开始h为0
    
    for i in range(step_num):
        x_history.append( x.copy() )

        grad = numerical_gradient(f, x)  # 计算梯度矢量
        h = h + grad * grad
        x = x - lr * grad / (np.sqrt(h) + 1e-7)

    return x, np.array(x_history)

init_x = np.array([-7.0, 2.0])  # 初始位置
lr = 0.9                        # 学习率
step_num = 50                   # 步数

x, x_history = AdaGrad(func, init_x, lr=lr, step_num=step_num)

step = 0.1
x = np.arange(-10, 10, step)
y = np.arange(-10, 10, step)
X, Y = np.meshgrid(x, y)  # 建立网格
Z = func_old(X, Y)

plt.figure(figsize=(10,6))
plt.contour(X, Y, Z, levels=50, linewidths=0.5, linestyles='dashdot')  # 绘制等高线

plt.plot(x_history[:,0], x_history[:,1], '-o')  # 绘制梯度下降过程

plt.xlim(-10, 10)
plt.ylim(-4, 4)
plt.xlabel("$x$")
plt.ylabel("$y$")

# 中文标题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('基于 AdaGrad 的最优化更新路径\n', size=15)

plt.show()
```

需要注意的是，我们在 AdaGrad 的实现中，在最后一行加上了微小值 1e-7，即 $1\times10^{-7}$，这是为了防止当 $h$ 中有 0 时，将 0 用作除数的情况。

> 在 PyTorch 中，AdaGrad 的实现是 `torch.optim.Adagrad`，这个微小值是可以调整的，默认为 1e-10。参考文档：[Adagrad](https://pytorch.org/docs/stable/optim.html?highlight=adagrad#torch.optim.Adagrad)。

从上图中可以看出，函数的取值高效地朝着最小值移动。由于 $y$ 轴方向上梯度较大，因此刚开始变动较大，但是后面会根据这个较大题变动按比例调整（参考前面的公式），减小更新的步伐。

因此，$y$ 轴方向上的更新程度被减弱。在本例中，“之”字形的变动程度明显减弱了。

### Adam

回顾一下之前的两种改进方法：

- Momentum：参照小球在碗中（山谷中）滚动的物理规则进行移动。
- AdaGrad：为每个元素适当地调整更新步伐。

如果将这两个方法融合在一起？ 这就是 **Adam** 方法的基本思路。

Adam 方法于 2015 年由论文 [Adam: A Method for Stochastic Optimization](https://arxiv.org/abs/1412.6980) 提出。

简单地说，就是融合了 Momentum 和 AdaGrad 的方法。通过组合两者的优点，有一定希望可以在参数空间中实现高效搜索。具体原理可以参考论文，在这里我们就不详细分析了。

```python
def Adam(f, init_x, lr=0.5, beta1=0.9, beta2=0.999, step_num=100):
    """
    Adam 方法
    
    f：要进行最优化的参数
    init_x：初始值
    lr：学习率，默认为0.5
    step_sum：梯度下降法重复的次数
    beta1：一阶矩估计的指数衰减率
    beta2：二阶矩估计的指数衰减率
    """
    x = init_x
    x_history = []  # 保存每一步的信息
    m = 0
    v = 0
    unbias_m = 0
    unbisa_b = 0
    
    for i in range(step_num):
        x_history.append( x.copy() )
        i = i + 1
        grad = numerical_gradient(f, x)                    # 计算梯度矢量
        lr_t = lr * np.sqrt(1.0 - beta2**i) / (1.0 - beta1**i)  
        m = m + (1.0 - beta1) * (grad - m)
        v = v + (1.0 - beta2) * (grad**2 - v)
        x = x - lr_t * m / (np.sqrt(v) + 1e-7)
        unbias_m += (1 - beta1) * (grad - m)               # correct bias
        unbisa_b += (1 - beta2) * (grad*grad - v)          # correct bias
        # x = x - lr * unbias_m / (np.sqrt(unbisa_b) + 1e-7)
        
    return x, np.array(x_history)

init_x = np.array([-7.0, 2.0])  # 初始位置
lr = 0.9                        # 学习率
step_num = 100                  # 步数
beta1 = 0.55
beta2 = 0.999
x, x_history = Adam(func, init_x, lr=lr, step_num=step_num, beta1=beta1, beta2=beta2)

step = 0.1
x = np.arange(-10, 10, step)
y = np.arange(-10, 10, step)
X, Y = np.meshgrid(x, y)  # 建立网格
Z = func_old(X, Y)

plt.figure(figsize=(10,6))
plt.contour(X, Y, Z, levels=50, linewidths=0.5, linestyles='dashdot')  # 绘制等高线

plt.plot(x_history[:,0], x_history[:,1], '-o')  # 绘制梯度下降过程

plt.xlim(-10, 10)
plt.ylim(-4, 4)
plt.xlabel("$x$")
plt.ylabel("$y$")

# 中文标题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('基于 Adam 的最优化更新路径\n', size=15)

plt.show()
```
